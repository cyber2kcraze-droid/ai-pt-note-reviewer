```python
# Add completeness score (0–1 based on Subjective, Objective, Assessment, Plan)
cols = ["has_subjective", "has_objective", "has_assessment", "has_plan"]
reviews_df["completeness_score"] = reviews_df[cols].sum(axis=1) / len(cols)

# Number of missing/vague elements (because it's a Python list right now)
reviews_df["num_missing_elements"] = reviews_df["missing_or_vague_elements"].apply(len)

# Has any red flags?
reviews_df["has_red_flags"] = reviews_df["red_flags_detected"].apply(lambda x: len(x) > 0)

# Simple overall score: completeness minus a penalty for missing elements
reviews_df["overall_score"] = reviews_df["completeness_score"] - 0.1 * reviews_df["num_missing_elements"]

reviews_df
```

```python
import pandas as pd

# Make a copy so we don't mess up your original
excel_df = reviews_df.copy()

# Convert list columns into readable strings for Excel
list_cols = ["missing_or_vague_elements", "red_flags_detected", "improvement_suggestions"]
for c in list_cols:
    if c in excel_df.columns:
        excel_df[c] = excel_df[c].apply(lambda x: "; ".join(x) if isinstance(x, list) else ("" if pd.isna(x) else str(x)))

# Make booleans consistent (TRUE/FALSE) for Excel readability
bool_cols = ["has_subjective", "has_objective", "has_assessment", "has_plan", "red_flags_addressed"]
for c in bool_cols:
    if c in excel_df.columns:
        excel_df[c] = excel_df[c].astype(bool)

excel_df.head()
```

```python
# Completeness score: 0–4 (or 0–1 if you want later)
parts = ["has_subjective", "has_objective", "has_assessment", "has_plan"]
excel_df["completeness_score"] = excel_df[parts].sum(axis=1)

# Count how many missing/vague items were found
excel_df["missing_count"] = excel_df["missing_or_vague_elements"].apply(lambda s: 0 if not s else len(str(s).split("; ")))

# Count red flags
excel_df["red_flag_count"] = excel_df["red_flags_detected"].apply(lambda s: 0 if not s else len(str(s).split("; ")))

# Simple overall label
def label_row(r):
    if r["red_flag_count"] > 0:
        return "Needs review (red flags)"
    if r["missing_count"] >= 3:
        return "Incomplete"
    if r["completeness_score"] == 4 and r["missing_count"] == 0:
        return "Strong"
    return "OK"

excel_df["quality_label"] = excel_df.apply(label_row, axis=1)

excel_df[["note_id", "completeness_score", "missing_count", "red_flag_count", "quality_label"]].head()
```
